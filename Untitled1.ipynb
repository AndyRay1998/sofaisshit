```python
import matplotlib.pyplot as plt
%matplotlib inline
from Brain_6cable_scene_gym import Brain_6cable_scene
import os
```

```python
import gym
from stable_baselines3 import PPO
from stable_baselines3.common.logger import configure

models_dir = "models/PPO"
logdir = "logs"
if not os.path.exists(models_dir):
    os.makedirs(models_dir)
    
if not os.path.exists(logdir):
    os.makedirs(logdir)
    
# set up logger
new_logger = configure(logdir, ["stdout", "csv"])

env = Brain_6cable_scene()
model = PPO("MlpPolicy", env, verbose=1, n_steps=1024, batch_size=128)
model.set_logger(new_logger)
model.learn(total_timesteps=2e5)
```

```python
for i in range(1,10):
    model.learn(total_timesteps=1e5,reset_num_timesteps=False, tb_log_name="PPO")
    model.save(f"{models_dir}/{i}")
```

```python
i = 2
model.load(f"{models_dir}/{i-1}")
model.learn(total_timesteps=1e5,reset_num_timesteps=False, tb_log_name="PPO")
model.save(f"{models_dir}/{i}")
```

```python
vec_env = model.get_env()
obs = vec_env.reset()
reward_l = []
for i in range(1000):
    action, _state = model.predict(obs, deterministic=True)
    obs, reward, done, info = vec_env.step(action)
    reward_l.append(reward)
```

```python
plt.plot(reward_l)
```
